## Exercício para os três tipos básicos de gráficos

Você recebeu, na forma de listas, os dados de receita anual de uma empresa de tecnologia, em bilhões de dólares, nos últimos anos. Com esses dados, você deve criar três gráficos, um de cada tipo abaixo listado, para visualizar esses dados. Os gráficos devem conter título e rótulos dos eixos. O nome da empresa é "Apple Inc.".

Tipos de gráficos:

- Gráfico de linhas
  - além da linha, mostre os pontos de cada valor
- Gráfico de barras
  - adicione o valor de cada barra no gráfico
- Gráfico de dispersão
  - ligue a grade do gráfico

Cada gráfico deve usar uma paleta de cores diferente.

Os dados são:

Ano | Receita (em bilhões de dólares)
--- | ---
| 2009 | 42,9
| 2010 | 65,2
| 2011 | 108,2
| 2012 | 156,5
| 2013 | 170,9
| 2014 | 182,8
| 2015 | 233,7
| 2016 | 215,6
| 2017 | 229,2
| 2018 | 265,6
| 2019 | 260,2
| 2020 | 274,5
| 2021 | 365,8
| 2022 | 394,3

Qual o tipo de gráfico mais adequado para visualizar esses dados?

### GRÁFICO DE LINHA

```python
# convertendo os dados para listas
ano = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
receita = [42.9, 65.2, 108.2, 156.5, 170.9, 182.8, 233.7, 215.6, 229.2, 265.6, 260.2, 274.5, 365.8, 394.3]

# gráfico de linhas
import matplotlib.pyplot as plt
from cycler import cycler

# cores no  gráfico
cores = plt.get_cmap('tab20b').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.figure(figsize=(15,5))
plt.plot(ano,receita,marker='o')
plt.title('Evolução da Receita Aple Inc.')
plt.ylabel('Receita em (US$)')
plt.xlabel('Anos')
plt.xticks(ticks= ano,labels=ano)
plt.show()
```

### GRÁFICO DE BARRA

```python
import matplotlib.pyplot as plt
from cycler import cycler



ano = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
receita = [42.9, 65.2, 108.2, 156.5, 170.9, 182.8, 233.7, 215.6, 229.2, 265.6, 260.2, 274.5, 365.8, 394.3]

#alterar tamanho figura grafico
# figsize=(largura,altura)
plt.figure(figsize=(15,5))

# cores do gráfico
cores = plt.get_cmap('Dark2').colors

# asssociar grafico a uma variavel: isto não modifica a contrução
# print(barras)
barras = plt.bar(ano,receita,color=cores[4])

# rotulo das barras
plt.bar_label(barras,labels=receita,padding=3,fontsize=10,fontweight='bold',color=cores[4])

#titulo e rotulos dos eixos
plt.title('Evolução da Receita Aple Inc.',fontweight='bold')
plt.ylabel('Receita(em milhões US$)',fontweight='bold')
plt.xlabel('Anos',fontweight='bold')

#garantir quetodos os anos aparecam e rotacao dos labels
plt.xticks(ano,rotation=45)
#mostrar grafico
plt.show()
```

### GRÁFICO DE DISPERSÃO

```python
import matplotlib.pyplot as plt
from cycler import cycler

ano = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
receita = [42.9, 65.2, 108.2, 156.5, 170.9, 182.8, 233.7, 215.6, 229.2, 265.6, 260.2, 274.5, 365.8, 394.3]

#criar figura
plt.figure(figsize=(15,5))

#cores do grafico
cores = plt.get_cmap('tab20b').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

# tamanho dos pontos 
# para mostrar em cor reversa: cmap='Reds_r'
plt.scatter(ano,receita,s=receita, c=receita,cmap='Reds')

plt.title('Evolução da Receita Aple Inc.',fontweight='bold')
plt.ylabel('US$')
plt.xlabel('Anos')
plt.xticks(ticks= ano,labels=ano)
#escala de cor 
plt.colorbar()

#linha de grade
plt.grid(True,linestyle='--')
plt.show()
```

### GRÁFICOS COMPILADOS

```python
import matplotlib.pyplot as plt
from cycler import cycler

# Dados
ano = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
receita = [42.9, 65.2, 108.2, 156.5, 170.9, 182.8, 233.7, 215.6, 229.2, 265.6, 260.2, 274.5, 365.8, 394.3]

# Criar figura e subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Mapa de cores
cmap = plt.get_cmap('RdYlGn')
norm = plt.Normalize(min(receita), max(receita))
cores = [cmap(norm(valor)) for valor in receita]

# Gráfico de Dispersão
axs[0].scatter(ano, receita, color=cores, s=100, edgecolor='black', alpha=0.7)
axs[0].set_title('Gráfico de Dispersão', fontsize=14, fontweight='bold')
axs[0].set_xlabel('Anos', fontsize=12)
axs[0].set_ylabel('Receita (US$ bilhões)', fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.7)

# Gráfico de Linha
axs[1].plot(ano, receita, marker='o', color=cores[0], linestyle='-', linewidth=2, markersize=8)
axs[1].set_title('Gráfico de Linha', fontsize=14, fontweight='bold')
axs[1].set_xlabel('Anos', fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.7)

# Gráfico de Barras
axs[2].bar(ano, receita, color=cores, edgecolor='black')
axs[2].set_title('Gráfico de Barras', fontsize=14, fontweight='bold')
axs[2].set_xlabel('Anos', fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.7)

# Ajustar layout
plt.suptitle('Evolução da Receita Apple Inc.', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Adicionar uma barra de cores para a legenda
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=axs, orientation='vertical', fraction=0.025, pad=0.02)
cbar.set_label('Receita (US$ bilhões)', fontsize=12)

# Mostrar gráfico
plt.show()
```
