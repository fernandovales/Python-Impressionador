```python
#dados
disciplinas =['banco de dados','comunicação de dados','empreededorismo','estrutura de dados','python','projeto integrado 2']
notas_ap1 =[10, 9.5, 8, 10, 10, 10]
notas_ap2 = [8, 10, 7, 9, 10, 10]
medias =[(notas_ap1[i] + notas_ap2[i])/2 for i in range(len(notas_ap1))]
```

```python
#ordenando dados
dados_ordenados = sorted(zip(medias,disciplinas),reverse=True)
medias_ordenadas = [dado[0] for dado in dados_ordenados]
disciplinas_ordenadas =[dado[1] for dado in dados_ordenados]
```

```python
#graficos
import matplotlib.pyplot as plt
from cycler import cycler

cores = plt.get_cmap('Pastel2').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.figure(figsize=(15,5))
barras = plt.bar(disciplinas_ordenadas, medias_ordenadas)

plt.bar_label(barras, labels=medias_ordenadas,
              padding=3, fontsize=10, fontweight='bold')

# retirando as bordas do gráfico
plt.box(False)

# retirando o eixo y
plt.yticks([])
plt.tick_params(axis='x', length=0)


# destacando a barra do produto com maior quantidade vendida
for i, barra in enumerate(barras):
    if i != medias_ordenadas.index(max(medias_ordenadas)):
        barras[i].set_color(plt.get_cmap('Pastel2').colors[1])
        barras[i].set_alpha(0.2)


plt.title('Médias do 3° Semestre - ADS')

# caso queira inverter o eixo, descomente a linha abaixo
# plt.gca().invert_xaxis()

plt.show()
```

```python
import matplotlib.pyplot as plt
from cycler import cycler

# Dados
disciplinas = ['Comunicação de Dados',
               'Estrutura de Dados', 
               'Python',
               'Banco de Dados',
               'Empreendedorismo',
               'Projeto Integrado']
notas = [9.75, 9.5, 10, 9, 7.5, 10]

# Ordenar as disciplinas e notas
disciplinas_ordenadas = [disciplina for _, disciplina in sorted(zip(notas, disciplinas))]
medias_ordenadas = sorted(notas)

# Criar um dicionário de cores para as notas
unique_notas = sorted(set(notas))
colors = plt.get_cmap('Pastel2').colors
nota_color_map = {nota: colors[i % len(colors)] for i, nota in enumerate(unique_notas)}

# Ajustar o ciclo de cores com base nas notas
ciclo_cores = [nota_color_map[nota] for nota in medias_ordenadas]
plt.rc('axes', prop_cycle=cycler('color', ciclo_cores))

plt.figure(figsize=(15, 5))
barras = plt.bar(disciplinas_ordenadas, medias_ordenadas, color=ciclo_cores)

plt.bar_label(barras, labels=medias_ordenadas, padding=3, fontsize=10, fontweight='bold')

# Retirando as bordas do gráfico
plt.box(False)

# Retirando o eixo y
plt.yticks([])
plt.tick_params(axis='x', length=0)

# Destacando a barra com a maior nota
maior_nota = max(medias_ordenadas)
for i, barra in enumerate(barras):
    if barra.get_height() == maior_nota:
        barra.set_color('green')
        barra.set_alpha(0.6)

plt.title('Médias do 3° Semestre - ADS')

# Caso queira inverter o eixo, descomente a linha abaixo
# plt.gca().invert_xaxis()

plt.show()
```
